```python
inv_pg = pd.read_excel(data_storage_loc + '/' + 'China_data.xlsx', sheet_name='Inventory-PG DC',skiprows = [0,1])
inv_pg.rename(columns={'SKU Code':'sku','Material Description':'material_description','CN Desc':'cn_desc',9264:'A9264','Material':'material','Unit':'unit'},inplace=True)
inv_pg = pd.melt(inv_pg,id_vars=['material','material_description','sku','cn_desc','unit'], value_vars=[ 'A9264','C816','A673','A672','A680','A668','A715','A716'],var_name='dc', value_name='inventory')
inv_pg['inventory'] = inv_pg['inventory'].fillna(0)
inv_pg.rename(columns={'dc':'warehouse'},inplace=True)
inv_pg['customer'] = 'ALL'
inv_pg.head()
```

```python

 
sheets_dict = pd.read_excel(data_storage_loc + '/' + 'China FAB shipment 0926.xlsx', sheet_name = None)
 
shipment_tmall_2020 = pd.DataFrame()
for name, sheet in sheets_dict.items():
    sheet['sheet'] = name
    sheet = sheet.rename(columns=lambda x: x.split('\n')[-1])
    shipment_tmall_2020 = shipment_tmall_2020.append(sheet)
    
shipment_tmall_2020.reset_index(inplace=True, drop=True)
shipment_tmall_2020 = shipment_tmall_2020[['material_available_date','ITBarcode','plant','Deliver(CS)','Deliver(MSU)','Deliver(GIV)','Deliver(PCS)','CaseCount']]
shipment_tmall_2020.rename(columns={'material_available_date':'date','ITBarcode':'sku','Deliver(CS)':'delivery_cases','Deliver(MSU)':'delivery_msu',
                               'Deliver(GIV)':'delivery_giv','Deliver(PCS)':'delivery_pcs','CaseCount':'case_count'},inplace=True)
shipment_tmall_2020 = shipment_tmall_2020[['date','sku','plant','delivery_cases','delivery_giv','delivery_pcs']]
shipment_tmall_2020 = shipment_tmall_2020[~shipment_tmall_2020['date'].isnull()]
shipment_tmall_2020.rename(columns={'plant':'warehouse'},inplace=True)
shipment_tmall_2020['customer'] = 'TMALL'
shipment_tmall_2020

shipment_tmall = pd.read_excel(data_storage_loc + '/' + 'Shipment_China.xlsx', sheet_name='Shipment - Tmall')
shipment_tmall = shipment_tmall[['Actual_Ship_Date','ITBarcode','plant','Deliver(CS)','Deliver(MSU)','Deliver(GIV)','Deliver(PCS)','CaseCount']]
shipment_tmall.rename(columns={'Actual_Ship_Date':'date','ITBarcode':'sku','Deliver(CS)':'delivery_cases','Deliver(MSU)':'delivery_msu',
                               'Deliver(GIV)':'delivery_giv','Deliver(PCS)':'delivery_pcs','CaseCount':'case_count'},inplace=True)
shipment_tmall = shipment_tmall[['date','sku','plant','delivery_cases','delivery_giv','delivery_pcs']]
shipment_tmall = shipment_tmall[~shipment_tmall['date'].isnull()]
shipment_tmall.rename(columns={'plant':'warehouse'},inplace=True)
shipment_tmall['customer'] = 'TMALL'

shipment_tmall = pd.concat([shipment_tmall_2020,shipment_tmall]).reset_index(drop = True)
shipment_tmall.head()
```

```python
offtake_pr = pd.read_excel(data_storage_loc + '/' + 'China_data.xlsx', sheet_name='2022.01-07 TM Offtake',nrows= 18)
offtake_pr.rename(columns={'SKU编码':'sku','SKU AVG Price':'sku_desc_cn'},inplace=True)
offtake_pr = offtake_pr.melt(id_vars=[ 'sku','sku_desc_cn'],var_name="date",value_name="avg_price")
offtake_pr = offtake_pr[~(offtake_pr['sku'].isnull())]
offtake_pr = offtake_pr[~(offtake_pr['avg_price'].isnull())]
offtake_pr['sku'] = offtake_pr['sku'].astype('int64')

offtake_vol = pd.read_excel(data_storage_loc + '/' + 'China_data.xlsx', sheet_name='2022.01-07 TM Offtake',skiprows=20)
offtake_vol.rename(columns={'SKU编码':'sku','SKU sold/pcs':'sku_desc_cn'},inplace=True)
offtake_vol = offtake_vol.melt(id_vars=[ 'sku','sku_desc_cn'],var_name="date",value_name="volume")
offtake_vol = offtake_vol[~(offtake_vol['sku'].isnull())]
offtake_vol = offtake_vol[~(offtake_vol['volume'].isnull())]
offtake_vol['sku'] = offtake_vol['sku'].astype('int64')
offtake_2022 = pd.merge(offtake_vol,offtake_pr[['sku','date','avg_price']],on=['sku','date'],how='outer')
offtake_2022.head()
```

```python
offtake_2021 = pd.read_excel(data_storage_loc + '/' + 'China_data.xlsx', sheet_name='2020.07-2021.12 TM Offtake',skiprows = [0])
offtake_2021.drop(columns={ '售出支数','售出GIV','实际售出金额','biz split'},inplace=True)
offtake_2021.rename(columns={'SKU Code':'sku','Volume':'sku_desc_cn','Avg Price':'avg_price','GIV':'giv','NOS':'nos'},inplace=True)
offtake_2021 = offtake_2021.melt(id_vars=[ 'sku','sku_desc_cn','capacity','giv','nos','avg_price'],var_name="date",value_name="volume")
offtake_2021 = offtake_2021[~(offtake_2021['sku'].isnull())]
offtake_2021['sku'] = offtake_2021['sku'].astype('int64')
offtake_2021_copy= offtake_2021.copy()
offtake_2021 = offtake_2021[['sku','date','volume','avg_price']]
offtake_2021 = pd.merge(offtake_2021,offtake_2022[['sku','sku_desc_cn']].drop_duplicates(),on=['sku'],how='left')
offtake_2021.head()
```

```python
offtake_tmall = pd.concat([offtake_2021,offtake_2022]).reset_index(drop=True)
offtake_tmall.drop(columns={'sku_desc_cn'},inplace=True)
offtake_tmall.rename(columns={'avg_price':'price'},inplace = True)
offtake_tmall['customer'] = 'TMALL'
offtake_tmall = offtake_tmall[['sku','customer','price','date','volume']]

offtake_tmall = offtake_tmall[offtake_tmall['sku'] != 815517026081]  # Discontinued product. Asked by customer to remove

offtake_tmall.head()
```

## ABC - XYZ

```python
index_var = 'sku'
date_var = 'date'
sales_var = 'offtake'
price_var = 'price'


abc_data = offtake_all[[index_var, date_var, sales_var,price_var]]
xyz_data = offtake_all[[index_var, date_var, sales_var,price_var]]

abc_data['revenue'] = abc_data[sales_var]*abc_data[price_var]
abc_data = abc_data.groupby([index_var], as_index = False).revenue.sum()

abc_data = abc_data.sort_values('revenue', ascending=False)
abc_data['rev_cumsum'] = abc_data['revenue'].cumsum()/abc_data['revenue'].sum()

abc_data['ABC'] = 'C'
abc_data.loc[abc_data['rev_cumsum']<=0.9, 'ABC'] = 'B'
abc_data.loc[abc_data['rev_cumsum']<=0.8, 'ABC'] = 'A'


xyz_data['week'] = xyz_data[date_var].dt.week
xyz_data = xyz_data.groupby([index_var, date_var], as_index  = False)[sales_var].sum()

xyz_data_calc=xyz_data.groupby(index_var,as_index=False)[sales_var].agg(['std','mean']).reset_index()
xyz_data_calc['CV'] = xyz_data_calc['std']/xyz_data_calc['mean']
xyz_data_calc.loc[xyz_data_calc['CV']<=1,'XYZ'] = 'Y'
xyz_data_calc.loc[xyz_data_calc['CV']<=0.5,'XYZ'] = 'X'
xyz_data_calc.loc[xyz_data_calc['CV']>1,'XYZ'] = 'Z'
abc_xyz= abc_data.merge(xyz_data_calc, how= 'left', on = [index_var])
abc_xyz['ABC_XYZ'] = abc_xyz['ABC']+abc_xyz['XYZ']
abc_xyz['revenue'] = abc_xyz['revenue'].round(2)
abc_xyz['rev_cumsum'] = abc_xyz['rev_cumsum'].round(2)
abc_xyz['std'] = abc_xyz['std'].round(2)
abc_xyz['mean'] = abc_xyz['mean'].round(2)
abc_xyz['CV'] = abc_xyz['CV'].round(2)
abc_xyz.head()
```

## Quartile division

```python
si_data_demand = offtake_final[offtake_final['offtake'] > 0]
si_grouped= si_data_demand.groupby(['sku','customer','date']).agg(total_sale=('offtake','sum')).reset_index()
# Calulating average and standard deviation

cv_data = si_grouped.groupby(['sku','customer']).agg(average=('total_sale','mean'),
                                                    sd=('total_sale','std')).reset_index()
## Calculating CV_squared

cv_data['cv_sqr'] = (cv_data['sd']/cv_data['average'])**2
cv_data
prod_by_date= si_data_demand.groupby(['sku','customer','date']).agg(count=('offtake','count')).reset_index()
prod_by_date['sku_cust'] = prod_by_date['sku'].astype(str) + "_" + prod_by_date['customer'].astype(str)

skus=prod_by_date.sku_cust.value_counts()


new_df= pd.DataFrame()
for i in range(len(skus.index)):
    a= prod_by_date[prod_by_date['sku_cust']==skus.index[i]]
    a['previous_date']=a['date'].shift(1)
    new_df=pd.concat([new_df,a],axis=0)
    
new_df.info()
```

```python
new_df['duration']=new_df['date']- new_df['previous_date']
new_df['Duration']=new_df['duration'].astype(str).str.replace('days','')
new_df['Duration']=pd.to_numeric(new_df['Duration'],errors='coerce')
## Calculating ADI

ADI = new_df.groupby(['sku','customer']).agg(ADI = ('Duration','mean')).reset_index()

adi_cv=pd.merge(ADI,cv_data)
adi_cv.head()

plt.figure(figsize=(15,3))
plt.hist(cv_data['cv_sqr'], bins = 20)
plt.legend()
plt.title("Average demand interval")
plt.ylabel("Occurance")
plt.xlabel("# Average demand interval")
plt.show()

plt.figure(figsize=(15,5))
plt.hist(ADI['ADI'], bins = 20)
plt.legend()
plt.title("Average demand interval")
plt.ylabel("Occurance")
plt.xlabel("# Average demand interval")
plt.show()

def category(df):
    a=0
    
    if((df['ADI']<=1.34) & (df['cv_sqr']<=0.49)):
        a='Smooth'
    if((df['ADI']>=1.34) & (df['cv_sqr']>=0.49)):  
        a='Lumpy'
    if((df['ADI']<1.34) & (df['cv_sqr']>0.49)):
        a='Erratic'
    if((df['ADI']>1.34) & (df['cv_sqr']<0.49)):
        a='Intermittent'
    return a


## Categorizing products based on their forcastability

adi_cv['category']=adi_cv.apply(category,axis=1)
adi_cv.loc[adi_cv['category'] == 0,'category'] = 'Erratic'
adi_cv.category.value_counts()
import seaborn as sns
## Visualizing the categories
adi_cv_daily = adi_cv
sns.scatterplot(x='cv_sqr',y='ADI',hue='category',data=adi_cv_daily)
```

```python

def get_events():
    """
    Hard coded events for now.
    Following Prophet's format
    """
    cny_promotion = pd.DataFrame({
        'holiday': 'cny',
        'ds': pd.to_datetime(['2019-02-5', '2020-01-25', '2021-02-12', '2022-02-01', '2023-01-22']),
        'lower_window': -7,
        'upper_window': 7,
        'type': 'festival',
    })

    # 3 weeks before actual date
    pre_cny_promotion = pd.DataFrame({
        'holiday': 'pre_cny',
        'ds': pd.to_datetime(['2019-01-14', '2019-12-30', '2021-01-18', '2022-01-10', '2023-01-01']),
        'lower_window': 0,
        'upper_window': 21,
        'type': 'promotion',
    })

    women_day_promotion = pd.DataFrame({
        'holiday': 'women day',
        'ds': pd.to_datetime(['2019-03-08', '2020-03-08', '2021-03-08', '2022-03-08', '2023-03-08']),
        'lower_window': 0,
        'upper_window': 0,
        'type': 'promotion_specific_date',
    })

    mother_day_promotion = pd.DataFrame({
        'holiday': 'mothers day',
        'ds': pd.to_datetime(['2019-05-12', '2020-05-10', '2021-05-09', '2022-05-08', '2023-05-14']),
        'lower_window': -7,
        'upper_window': 0,
        'type': 'promotion_specific_date',
    })

    day_520_promotion = pd.DataFrame({
        'holiday': '520',
        'ds': pd.to_datetime(['2019-05-20', '2020-05-20', '2021-05-20', '2022-05-20', '2023-05-20']),
        'lower_window': 0,
        'upper_window': 0,
        'type': 'promotion_specific_date',
    })

    day_618_promotion = pd.DataFrame({
        'holiday': '618',
        'ds': pd.to_datetime(['2019-06-18', '2020-06-18', '2021-06-18', '2022-06-18', '2023-06-18']),
        'lower_window': -14,
        'upper_window': 14,
        'type': 'promotion_specific_date',
    })

    day_99_promotion = pd.DataFrame({
        'holiday': '9.9',
        'ds': pd.to_datetime(['2019-09-09', '2020-09-09', '2021-09-09', '2022-09-09', '2023-09-09']),
        'lower_window': -7,
        'upper_window': 7,
        'type': 'promotion_specific_date',
    })

    day_1111_promotion = pd.DataFrame({
        'holiday': '11.11',
        'ds': pd.to_datetime(['2019-11-11', '2020-11-11', '2021-11-11', '2022-11-11', '2023-11-11']),
        'lower_window': -14,
        'upper_window': 7,
        'type': 'promotion_specific_date',
    })

    day_1212_promotion = pd.DataFrame({
        'holiday': '12.12',
        'ds': pd.to_datetime(['2019-12-12', '2020-12-12', '2021-12-12', '2022-12-12', '2023-12-12']),
        'lower_window': -7,
        'upper_window': 7,
        'type': 'promotion_specific_date',
    })

    black_friday_promotion = pd.DataFrame({
        'holiday': 'black_friday',
        'ds': pd.to_datetime(['2019-11-29', '2020-11-27', '2021-11-26', '2022-11-25', '2023-11-24']),
        'lower_window': -7,
        'upper_window': 7,
        'type': 'promotion_specific_date',
    })

    christmas_promotion = pd.DataFrame({
        'holiday': 'christmas',
        'ds': pd.to_datetime(['2019-12-25', '2020-12-25', '2021-12-25', '2022-12-25', '2023-12-25']),
        'lower_window': -7,
        'upper_window': 7,
        'type': 'festival',
    })

    events = pd.concat((
        pre_cny_promotion,
        cny_promotion,
        women_day_promotion,
        mother_day_promotion,
        day_520_promotion,
        day_618_promotion,
        day_99_promotion,
        day_1111_promotion,
        day_1212_promotion,
        black_friday_promotion,
        christmas_promotion,
    )).reset_index(drop=True)

    a = pd.get_dummies(events["type"])
    events = pd.concat([events, a], axis=1)
    return events

events = get_events()
events['upper_window']  = events['upper_window'].astype(int)
events['lower_window']  = events['lower_window'].astype(int)
events['festival']  = events['festival'].astype(int)
events['promotion']  = events['promotion'].astype(int)
events['promotion_specific_date']  = events['promotion_specific_date'].astype(int)
events_weekly = []
for h in events.holiday.unique():
    # for each event, resample it to W-Mon
    e = events.loc[events.holiday == h]
    e = e.set_index("ds").resample("W-Mon", label="left").first()
    e.dropna(inplace=True)
    events_weekly.append(e.reset_index())
event_types = pd.get_dummies(events["type"]).columns.to_list()
events_weekly = pd.concat(events_weekly)
tmp = []
for t in events_weekly.ds.unique():
    e = events_weekly.loc[events_weekly.ds == t].reset_index(drop=True)
    if len(e) == 1:
        tmp.append(e.drop(columns=["lower_window", "upper_window"]))
    else:
        e.drop(columns=["lower_window", "upper_window"], inplace=True)
        e["holiday"] = ",".join(e["holiday"].tolist())
        e["type"] = ",".join(e["type"].tolist())
        _sum = e[event_types].sum()
        for c in event_types:
            e[c] = _sum[c]
        tmp.append(e.drop_duplicates())
events_weekly = pd.concat(tmp)
events_weekly.sort_values(by="ds", inplace=True)
events_weekly.reset_index(inplace=True, drop=True)
```
